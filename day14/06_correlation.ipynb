# Correlation

- What is Correlation?
- Variables within a dataset can be related for lots of reasons
- Types:
  - Pearson's r
  - Spearman's rho
  - Kendall's tau

For example:

1. One variable could cause or depend on the values of another variable.
2. One variable could be lightly associated with another variable.
3. Two variables could depend on a third unknown variable.

**Positive Correlation:** Both variables change in the same direction.

**Neutral/Zero Correlation:** No relationship in the change of the variables.

**Negative Correlation:** Variables change in opposite directions/

# Covariance

- Variables can be related by linear relationship.
- This relationship can be summarized between two variables, called Covariance.
- The sign of the covariance can be interpreted as whether the two variables change.
- The magitude of covariance is not easily interpreted.

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset
ship = sns.load_dataset('titanic')
flower = sns.load_dataset('iris')
```

```python
ship.head()
```

```python
# both numerical values
np.cov(ship['age'], ship['fare']) # covariance
```

```python
# Python code to demonstrate the
# use of numpy.cov
import numpy as np

x = [1.23, 2.12, 3.34, 4.5]

y = [2.56, 2.89, 3.76, 3.95]

# find out covariance with respect columns
cov_mat = np.stack((x, y), axis = 0)
cov_mat

print(np.cov(cov_mat))
```

```python
# Calulate pearsonsr correlation
from scipy.stats import pearsonr
corr, _ = pearsonr(flower['sepal_length'], flower['petal_length'])
print('Pearsons Correlation: %.3f' % corr)
```

```python
# Calulate spearmanr correlation
from scipy.stats import spearmanr
corr, _ = spearmanr(flower['sepal_length'], flower['petal_length'])
print('Spearmanr Correlation: %.3f' % corr)
```
