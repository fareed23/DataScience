```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.info()
```

```python
# To check for null values
df.isnull().sum()
```

```python
# To plot in other way
df.survived.value_counts().plot(kind="bar")
```

```python
# Bar plot
sns.countplot(x='survived', hue='sex', data=df)
```

```python
sns.barplot(data=df, x='sex', y='fare', hue="pclass")
```

```python
# Scatter plot 
plt.scatter(data=df, x='age', y='fare', cmap="pclass") # To check if the agr depends on age
```

```python
# Scatter plot -> should be both numeric
sns.scatterplot(data=df, x='age', y='fare', hue='pclass') # To check if the agr depends on age
```

```python
# Pair plot
sns.pairplot(df)
```

```python
# Histogram
df.hist(bins=6, figsize=(8,6), grid=False)
```

```python
sns.boxplot(data=df, x='sex', y='fare', hue='class')
```

```python
sns.boxenplot(data=df, x='sex', y='fare', hue='who')
```

```python
# Grids
g = sns.FacetGrid(df, hue='survived', col='pclass', margin_titles=True, palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend()
```

```python
g = sns.FacetGrid(df, hue='survived', col='sex', margin_titles=True, palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend()
```

```python
# Strip plot
sns.stripplot(data=df, x='sex', y='fare', hue='class')
```

```python
# Heat Map
corr = df.corr() # corr -> Correlation
plt.figure(figsize=(8, 8))
sns.heatmap(corr)
plt.title("Heat Map of Titanic")
```

```python
sns.violinplot(data=df, x='sex', y='fare', hue='class').set_title("Violin Plot", fontdict={'size':30, 'weight':'bold', 'color':'gray'})

plt.show()
```

```python
df = sns.load_dataset('tips')
chart = sns.barplot(data=df, x='day', y='tip', ci=None)
chart.set_title('Sample Title', fontdict={'size': 30, 'weight': 'bold'})
chart.set_xlabel('Day of Week', fontdict={'size': 15})
chart.set_ylabel('Tip Amount', fontdict={'size': 15})
```

```python
# Adding Labels in an Easier Way

df = sns.load_dataset('tips')
sns.barplot(data=df, x='day', y='tip', ci=None).set(title='Sample Title', xlabel='Day', ylabel='Tip')
plt.show()
```
