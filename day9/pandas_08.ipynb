```python
# Import dataset
df = pd.read_excel('titanic.xlsx')
```

```python
# To get any column by it's name
# if we don't know any column names
df.columns
```

```python
# And then if on column we put -> 1D [] (one column)
df['embark_town']
```

```python
# And if we want two or more columns we put 2D [[]] (column and rows both)
df[['embark_town', 'survived', ]]
```

```python
# To delete a column
# df.drop(columns='Unnamed: 0')
df = df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
df.groupby(['sex']).describe().T
```

```python
df.count()
```

```python
# Count the numbers of survived = 1 grouped by sex (male or female)
df.groupby(['sex']).count()
```

```python
df.groupby(['survived', 'sex']).count()
```

```python
df.groupby(['survived', 'sex'])['survived'].count()
```

```python
df.groupby(['survived', 'sex', 'who'])['who'].count()
```

```python
# Now let's create graph
sns.countplot(x='survived', data=df)
```

```python
# hue for colors detailed
sns.countplot(x='survived', hue='sex' ,data=df)
```
